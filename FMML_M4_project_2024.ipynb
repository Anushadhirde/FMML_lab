### Gradient Descent

````python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

def df(x):
    # Example slope function: df/dx = 2x (Modify as per your requirement)
    return 2 * x

def f(x):
    # Example cost function: f(x) = x^2 (Modify as per your requirement)
    return x**2

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED]
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED]

    step_size = abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED]

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
## TASK-5 : Run gradient descent 3 times
## TASK-6 : Plot two more learning rates: mid_gamma (0.001) and high_gamma (0.002)
# Gradient Descent Function
def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    steps = [x]

    for i in range(max_iter):
        gradient = derivative_func(x)
        new_x = x - multiplier * gradient

        steps.append(new_x)

        if abs(new_x - x) < precision:
            break

        x = new_x

    return steps

# Define parameters
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.001, precision=0.0001, max_iter=n)  # [REQUIRED]

high_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                              multiplier=0.002, precision=0.0001, max_iter=n)  # [REQUIRED]

insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9,
                                multiplier=0.25, precision=0.0001, max_iter=n)

# Plot the results
plt.figure(figsize=(12, 8))

# Plot Low Gamma
plt.plot(low_gamma, label='Low Gamma (0.0005)', marker='o', linestyle='--')

# Plot Mid Gamma
plt.plot(mid_gamma, label='Mid Gamma (0.001)', marker='o', linestyle='--')

# Plot High Gamma
plt.plot(high_gamma, label='High Gamma (0.002)', marker='o', linestyle='--')

# Plot Insane Gamma
plt.plot(insane_gamma, label='Insane Gamma (0.25)', marker='o', linestyle='--')

# Customize the plot
plt.title("Gradient Descent with Different Learning Rates")
plt.xlabel("Iteration")
plt.ylabel("x value")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--', label="x=0 (Target)")
plt.legend()
plt.grid()

plt.show()
````

````python
## TASK-7 : Plot reduction in cost for each iteration
# Gradient Descent Function (Modified to include costs)
def gradient_descent_with_cost(derivative_func, cost_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    steps = [x]
    costs = [cost_func(x)]  # Track the cost function value at each step

    for i in range(max_iter):
        gradient = derivative_func(x)
        new_x = x - multiplier * gradient
        steps.append(new_x)
        costs.append(cost_func(new_x))  # Compute and store the new cost

        if abs(new_x - x) < precision:
            break

        x = new_x

    return steps, costs

# Parameters
n = 100
low_gamma_steps, low_gamma_costs = gradient_descent_with_cost(derivative_func=dg, cost_func=g,
                                                              initial_guess=3, multiplier=0.0005,
                                                              precision=0.0001, max_iter=n)

mid_gamma_steps, mid_gamma_costs = gradient_descent_with_cost(derivative_func=dg, cost_func=g,
                                                              initial_guess=3, multiplier=0.001,
                                                              precision=0.0001, max_iter=n)

high_gamma_steps, high_gamma_costs = gradient_descent_with_cost(derivative_func=dg, cost_func=g,
                                                                initial_guess=3, multiplier=0.002,
                                                                precision=0.0001, max_iter=n)

# Plot the reduction in cost
plt.figure(figsize=(12, 8))

# Plot for Low Gamma
plt.plot(low_gamma_costs, label='Low Gamma (0.0005)', marker='o', linestyle='--')

# Plot for Mid Gamma
plt.plot(mid_gamma_costs, label='Mid Gamma (0.001)', marker='o', linestyle='--')

# Plot for High Gamma
plt.plot(high_gamma_costs, label='High Gamma (0.002)', marker='o', linestyle='--')

# Customize the plot
plt.title("Reduction in Cost Function Across Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost Function g(x)")
plt.legend()
plt.grid()

plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
import numpy as np

# Gradient of the cost function
def grad(x, y, thetas):
    n = y.size

    # Calculate the partial derivatives for theta0 and theta1
    theta0_slope = (-2/n) * np.sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * np.sum((y - thetas[0] - thetas[1]*x) * x)

    # Return the gradients as a 1D array
    return np.array([theta0_slope, theta1_slope])

# Example usage:
x = np.array([1, 2, 3, 4])
y = np.array([3, -0.5, 2, 7])
thetas = np.array([0.5, 0.5])  # Initial theta values

# Call the grad function
gradients = grad(x, y, thetas)
print(f'Gradients: {gradients}')
````
